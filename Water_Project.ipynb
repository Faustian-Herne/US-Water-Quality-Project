```python
ars = pd.read_csv('us_arsenic_water_samples_2020_2024.csv', low_memory=False)
```

```python
ars.columns.tolist()
```

```python
ars.head()
```

```python
ars1 = ars[['CharacteristicName','ResultSampleFractionText', 'ActivityStartDate','ResultMeasureValue','ResultMeasure/MeasureUnitCode','ActivityLocation/LatitudeMeasure','ActivityLocation/LongitudeMeasure']]
```

```python
ars1 = ars1[ars1['ResultMeasureValue'] != '_']
```

```python
ars1 = ars1[ars1['ActivityLocation/LatitudeMeasure'].notnull()]
```

```python
ars1 = ars1[ars1['ResultMeasureValue'].notnull()]
```

```python
ars1 = ars1[ars1['ResultMeasure/MeasureUnitCode'].notnull()]
```

```python
ars1.head()
```

```python
ars1 = ars1[~ars1.ResultMeasureValue.isin(['ND','Not Reported'])]
```

```python
ars1['ResultMeasure/MeasureUnitCode'] = ars1['ResultMeasure/MeasureUnitCode'].str.replace('mg/kg', 'ug/L').str.replace('ppb', 'ug/L').str.replace('ppm', 'mg/L')
```

```python
ars1['ResultMeasureValue'] = ars1['ResultMeasureValue'].str.replace('<', '')
```

```python
ars1 = ars1[ars1['ResultMeasureValue'].notnull()]
```

```python
ars1['ResultMeasureValue'] = ars1['ResultMeasureValue'].astype(float, errors='ignore')
```

```python
condition = (ars1['ResultMeasure/MeasureUnitCode'] == 'mg/L')
ars1['ResultMeasureValue'] = np.where(condition, ars1['ResultMeasureValue'] * 1000, ars1['ResultMeasureValue'])
ars1['ResultMeasure/MeasureUnitCode'] = ars1['ResultMeasure/MeasureUnitCode'].str.replace('mg/L', 'ug/L')
condition = (ars1['ResultMeasure/MeasureUnitCode'] == 'ng/L')
ars1['ResultMeasureValue'] = np.where(condition, ars1['ResultMeasureValue'] * 0.001, ars1['ResultMeasureValue'])
ars1['ResultMeasure/MeasureUnitCode'] = ars1['ResultMeasure/MeasureUnitCode'].str.replace('ng/L', 'ug/L')
```

```python
ars1.to_csv('us_arsenic_water_samples_abbr_2020_2024.csv', index=False)  
```
